# Edge Detection

```python
#resize image
resized_img = cv2.resize(img, (0,0), fx=0.8, fy=0.8)

cv2.imshow('show image',resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Apply Canny
gray = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (5,5), 0)
canny = cv2.Canny(blur, threshold1 = 100, threshold2 = 100)

cv2.imshow('show image',canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Working with video capture 

```python
count = 1

#establish capture
cap = cv2.VideoCapture('D:\\coding\\computer vision\\videos\\2022-08-21 排長龍睇醫生 covid.mp4')

#properties
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = cap.get(cv2.CAP_PROP_FPS)
# video writer
video_writer = cv2.VideoWriter('D:\\coding\\computer vision\\videos\\2022-08-21 排長龍睇醫生 covid_edited.mp4', cv2.VideoWriter_fourcc('P','I','M','1'), fps, (width, height))

#Loop through each frame
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    
    #read frame
    ret, frame = cap.read()

    #dynamic colouring
    #count += 1
    #if count % 10 == 0:
    #    dynamic_colour = cv2.cvtColor(frame, cv2.COLOR_BGR2HLS)
    #elif count % 7 == 0:
    #    dynamic_colour = cv2.cvtColor(frame, cv2.COLOR_BGR2BGRA)
    #elif count % 2 == 0:
    #    dynamic_colour = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    #elif count % 1 == 0:
    #    dynamic_colour = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    #canny edge all frames
    canny = cv2.Canny(frame, threshold1 = 180, threshold2 = 200)

    #show image
    cv2.imshow('Video Player', canny)
    # Write out frame
    video_writer.write(canny)
    
    #break loop
    if cv2.waitKey(20) & 0xFF == ord('q'):
         break

#close down everything
cap.release()
cv2.destroyAllWindows()
# release video writer
video_writer.release()
```
