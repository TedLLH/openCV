```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

# Scaling

```python
img = cv.imread('../images/flare.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
#by defaul, interpolation = cv.INTER_LINEAR 
#for shirnking = image decimation = subsampling = skipping, cv.INTER_AREA, gives moire'-free result
#for zooming, cv.INTER_CUBIC & cv.INTER_LINEAR
res = cv.resize(img,None,fx=0.5, fy=0.5, interpolation = cv.INTER_AREA)
cv.imshow('img',res)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#scale the smaller image to the size of th bigger image, add padding and maintain aspect ration
A = cv.imread('../images/flare.jpg')
B = cv.imread('../images/bike.jpg')

#size of the bigger image
width = A.shape[0]
height = A.shape[1]

# Pad smaller img
width_pad = int(abs((A.shape[0]-B.shape[0])/2))
height_pad = int(abs((A.shape[1]-B.shape[1])/2))
padded = cv.copyMakeBorder(B,width_pad,width_pad,height_pad,height_pad,cv.BORDER_CONSTANT) 

# Resize  
dsize = (height,width)
img1_resized = cv.resize(A, dsize)
img2_resized = cv.resize(padded, dsize)

cv.imshow('A', A)
cv.imshow('B', B)
cv.imshow('img1_resized', img1_resized)
cv.imshow('img2_resized', img2_resized)
cv.waitKey()
cv.destroyAllWindows()
```

# Shift an image

```python
img = cv.imread('../images/flare.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols = img.shape
#determine the tx and ty for the image position
M = np.float32([[1,0,500],[0,1,50]])
#3rd argument is size of the images, "width,height" == "cols,rows"
dst = cv.warpAffine(img,M,(cols,rows))
cv.imshow('img',dst)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Rotate an image

```python
img = cv.imread('../images/flare.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols = img.shape
# cols-1 and rows-1 are the coordinate limits.
M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),0,1)
dst = cv.warpAffine(img,M,(cols,rows))
cv.imshow('img',dst)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Affine transformation
All parallel lines willstill be parallel after the transformation

```python
img = cv.imread('../images/flare.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv.getAffineTransform(pts1,pts2)
dst = cv.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

# Perspective Transformation

```python
img = cv.imread('../images/flare.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,ch = img.shape
pts1 = np.float32([[250,200],[750,200],[750,800],[250,800]])
pts2 = np.float32([[0,0],[500,0],[500,600],[0,600]])
M = cv.getPerspectiveTransform(pts1,pts2)
dst = cv.warpPerspective(img,M,(500,600))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
